# Simulation réseau de pixels – Théorie ZM
Génération d’un graphe fractal de pixels quantiques et émergence de la métrique.

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Paramètres
N = 1000
p = 0.01
G = nx.erdos_renyi_graph(N, p)

# Attributs
for i in G.nodes:
    G.nodes[i]['rho'] = np.random.lognormal(0, 0.1)
    G.nodes[i]['theta'] = np.random.uniform(0, 2*np.pi)

# Visualisation
pos = nx.spring_layout(G, seed=42)
rho = [G.nodes[i]['rho'] for i in G.nodes]
nx.draw(G, pos, node_color=rho, cmap='plasma', node_size=10)
plt.title("Réseau de pixels ZM – densité informationnelle")
plt.savefig("pixel_network.png", dpi=300, bbox_inches="tight")
plt.show()
```
